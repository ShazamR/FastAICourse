```
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```
heights = np.array([147.9, 163.5, 159.8, 155.1, 163.3, 158.7, 172.0, 161.2, 153.9, 161.6])
weights = np.array([41.7, 60.2, 47.0, 53.2, 48.3, 55.2, 58.5, 49.0, 46.7, 52.5])

X = pd.DataFrame(heights, columns = ["Heights"])
target = pd.DataFrame(weights, columns = ["Weights"])
y = target["Weights"]

print(X)
print(y)
lr = LinearRegression()
lr.fit(X, y)

new_heights = pd.DataFrame(np.array([160, 180, 200]))
predicting_weights = lr.predict(new_heights)
print(predicting_weights)

plt.scatter(heights, weights)
trendline_weights = lr.predict(X)
plt.plot(heights, trendline_weights, color = "blue")
plt.plot(new_heights, predicting_weights, color = "red", marker = "o", markersize = 10)
plt.title("Predicting student weights by heights")
plt.xlabel("Heights")
plt.ylabel("Weights")
plt.savefig('heightVSweight.png')
plt.show()

```
