```
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
temperatures = np.array([29, 28, 34, 31, 25, 29, 32, 31, 24, 33, 25, 31, 26, 30])
drink_sales = np.array([7.7, 6.2, 9.3, 8.4, 5.9, 6.4, 8.0, 7.5, 5.8, 9.1, 5.1, 7.3, 6.5, 8.4])
X = pd.DataFrame(temperatures, columns=["Temperature"])
target = pd.DataFrame(drink_sales, columns = ["Drink_sales"])
y = target["Drink_sales"]
```

```
print(y)
```

```
print(target)
```

```
print(drink_sales)
```

```
lr = LinearRegression()
lr.fit(X, y)
```

```
new_temp = pd.DataFrame(np.array([26, 30]))
predicting_price = lr.predict(new_temp)
print(predicting_price)
```

```
plt.scatter(temperatures, drink_sales)
regression_sales = lr.predict(X)
plt.plot(temperatures, regression_sales, color = "blue")
plt.plot(new_temp, predicting_price, color = "red", marker = "o", markersize = 10)
plt.title("temperature vs price", size = 20, color = "white")
plt.xlabel("Temperature", color = "white")
plt.ylabel("Sales", color = "white")
plt.show()
```
